```python
# Handle missing values: 'director' is ~30% null, 'cast' is ~9% null [2, 3]
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration']= df['duration'].fillna('Unknown')
```

Q:what is the mean duriation in movies.

```python

movie = df[df["type"] == "Movie"].copy()
movie = movie[movie['duration'] != 'Unknown']
movie['duration'] = movie['duration'].str.replace(' min', '').astype(float)
print("mean duration in movies:\n", movie['duration'].mean().round(2))
```

Q:what is the mean duriation in tv shows.

```python
tvshow = df[df["type"] == "TV Show"].copy()
tvshow = tvshow[tvshow['duration'] != 'Unknown']
tvshow['duration'] = tvshow['duration'].str.replace(' Season', '').str.replace('s', '').astype(float)
print("mean duration in TV shows:\n", round(tvshow['duration'].mean(), 2))
```

Q:what are the top 10 movies in netflix with their ratings,which type of genere have more distribution,what are the top 10 TV-MA movies

```python
movies = df[df['type'] == 'Movie']
print("Names and Ratings of 10 Movies:")
print(movies[['title', 'rating']].head(10))
movies["rating"].value_counts().plot(kind="bar",figsize=(10,5),title="Movie rating distribution")
#optional to find top movies with any  rating.
tv_ma_movies = movies[movies['rating'] == 'TV-MA']
print("\nNames of 10 Movies rated TV-MA:")
print(tv_ma_movies[['title', 'rating']].head(10))
# all time top 10 movies in netflix with their ratings.
```

Q:Q:what are the top 10 tv shows in netflix with their ratings,which type of genere have more distribution,what are the top 10 TV-MA movies

```python
tvshows = df[df['type'] == 'TV Show']
print("\nNames and Ratings of 10 TV Shows:")
print(tvshows[['title', 'rating']].head(10))
tvshows[['title', 'rating']].head(10).set_index('title')['rating'].value_counts().plot(kind="bar",figsize=(10,5),title="Top 10 TV shows with their ratings")
#optional to find top TV shows with any  rating.
shows=tvshows[tvshows['rating'] == 'TV-MA']
print("\nNames of 10 TV Shows rated TV-MA:")            
print(shows[['title', 'rating']].head(10))
# all time top 10 TV shows in netflix with their ratings.
```
